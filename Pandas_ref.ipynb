#### Important Features of pandas 

- **Data Structures**: Provides powerful, flexible, and labeled data structures like Series and DataFrame for data manipulation.    
- **Data Cleaning**: Supports handling missing data with methods like `fillna()` and `dropna()`.  
- **Indexing and Slicing**: Offers robust label-based and position-based indexing.  
- **Data Aggregation**: Includes built-in functions for group operations via `groupby()`.  
- **Data Transformation**: Facilitates reshaping, pivoting of datasets.  
- **Time Series Handling**: Built-in support for time-series data with date-range generation and frequency conversion.  
- **Data Visualization**: Integrates with Matplotlib for quick plotting of data.  
- **File I/O**: Supports reading and writing data in multiple formats, including CSV, Excel, SQL, JSON, and more.    
- **Integration**: Works seamlessly with other libraries like NumPy, SciPy, and Scikit-learn.

#### Comparison between Pandas **Series** and **DataFrame**:

| Feature                     | **Pandas Series**                                  | **Pandas DataFrame**                           |
|-----------------------------|---------------------------------------------------|-----------------------------------------------|
| **Definition**              | A one-dimensional labeled array.                 | A two-dimensional labeled data structure.     |
| **Data Structure**          | Single column or array-like structure.           | Tabular structure with rows and columns.      |
| **Dimension**               | 1D (One-dimensional).                            | 2D (Two-dimensional).                         |
| **Labels**                  | Index labels for rows.                           | Index labels for rows and column labels.      |
| **Data Type**               | Homogeneous (all elements of the same type).     | Heterogeneous (different types per column).   |
| **Usage**                   | Used for single-variable datasets.               | Used for multi-variable datasets.             |
| **Creation**                | Created from lists, arrays, or dictionaries.     | Created from dictionaries, arrays, Series, etc. |
| **Access**                  | Accessed via index labels (e.g., `series[0]`).   | Accessed via row and column labels (e.g., `df.loc[]`, `df.iloc[]`). |
| **Example Structure**       | `[10, 20, 30]` with index `[0, 1, 2]`.           | Table-like structure with rows and columns.   |

#### Create pandas series

```python
# Import pandas
import pandas as pd
import numpy as np
```

```python
# Create pandas series using list

data = [10, 20, 20, 40,9,6.7]
series = pd.Series(data)
print(series)
print(series.dtype)
```

```python
# Create a series using a custom index/label
data = [10,20,30,40,12,80]  # list of values
i = ['a','b','c','d','e','f'] # index labels
series = pd.Series(data, index=i)
print(series)
```

```python
# Create pandas series using dictionary
data = {'a': 10, 'b': 20, 'c': 30,'d':5}
series = pd.Series(data)
print(series)
```

```python
series.replace(20,100)
```

```python
# Create series using numpy array
import numpy as np

data = np.array([100, 200, 300,50,400,75])
i = ['a','b','c','d','e','f'] # define labels
series = pd.Series(data, index=i)
print(series)
```

#### Pandas dataframe

```python
# Create pandas dataframe from dictionary of lists
import pandas as pd

data = {'Name': ['Alice', 'Bob','Mary',"John"], \
        'Age': [25, 30,34,56], \
        'Country':["USA","UK","AUS","India"]}
i = ['a','b','c','d']
df = pd.DataFrame(data, index=i)
print(df)
```

```python
# Create pandas dataframe from a 2D numpy array
import numpy as np

array = np.array([[10,5,20], [14, 25, 6],[7,18,29]])
print(array)
df = pd.DataFrame(array) # without column names
print(df)
```

```python
# Create pandas dataframe from a 2D numpy array
import numpy as np

array = np.array([[10,5,20], [14, 25, 6],[7,18,29]])
print(array)
df = pd.DataFrame(array, columns=["C1","C2","C3"]) \
     # with column names
print(df)
```

```python
# Create a dataframe from a nested list
data = [[1, 'Alice', 25], [2, 'Bob', 30],[3,'Mary',34]]
print(data)
df = pd.DataFrame(data, columns=['ID', 'Name', 'Age'])
print(df)
```

```python
# Create a dataframe with the array
data = {"weight":[66,75,84,96,48,71], \
      "height":[156,165,186,165,174,167]}
bmi = pd.DataFrame(data)
bmi
```

```python
# Create a new BMI column
bmi["BMI"] = bmi["weight"]/(bmi["height"]/100)**2
bmi
```

```python
#Insert new record
bmi.loc[3] = [78,np.nan,np.nan]
bmi
```

```python
# Insert another row with missing values
bmi.loc[5] = [np.nan, 65, np.nan]
bmi
```

```python
# Info()
bmi.info()
```

```python
# Check Column-wise sum of NaN values(missing values)
bmi.isnull().sum()
```

```python
# Replace missing value with 65kg value of that column
bmi["weight"] = bmi["weight"].fillna(65)
bmi
```

```python
# Replace missing value with 65kg value of that column
bmi["height"] = bmi["height"].fillna(165)
bmi
```

```python
bmi["BMI_New"] = bmi["weight"]/(bmi["height"]/100)**2
bmi
```

```python
# Drop BMI column permanently
bmi.drop("BMI", axis = 1, inplace=True)
bmi
```

#### Data Visualization on iris dataset

```python
iris_df = pd.read_excel("iris.xlsx")
```

```python
import matplotlib.pyplot as plt
# plot histogram on selected columns
plt.hist(iris_df["sepal.length"], color="blue")
```

```python
# plot histogram using seaborn
import seaborn as sns
plt.figure(figsize=(5,3))
plt.title("Histogram")
plt.xlabel("Sepal length")
plt.ylabel("Count")
sns.histplot(iris_df["sepal.length"], kde = True)
```

```python
# Box and Whisker's plot
plt.figure(figsize=(6,2))
plt.title("Box and whiskers plot")
plt.xlabel("Sepal length")
sns.boxplot(iris_df["sepal.length"],orient= 'h')
```

```python
# Use groupby variety
iris_df.groupby(iris_df["variety"]).mean()
```

```python
iris_df["sepal.length"].groupby(iris_df["variety"]).mean()
```

```python
fruits = ["Apple","Banana","Orange","Grapes"]
students = [10,8,7,5]

df = pd.DataFrame({"Fruit":fruits,"Students":students})
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(fruits,students,color=["red","blue","orange","green"],ec="black")
plt.xlabel("Fruits")
plt.ylabel("Students")
plt.show()
```

```python
plt.barh(fruits,students,color=["red","blue","orange","green"],ec="black")
plt.xlabel("Fruits")
plt.ylabel("Students")
plt.show()
```

```python
df2 = pd.DataFrame({
        "Fruits" : ["Apple","Banana","Orange","Grapes"],
        "Girls": [6,4,3,2],
        "Boys": [4,4,4,3]
})

df2   
```

```python
df2.plot(kind="bar",x="Fruits",ec="black")
plt.title("Grouped Bar Chart")
plt.legend()
plt.ylabel("No.of Students")
plt.xticks(rotation=45,ha="right")
plt.show()
```

```python
df2.plot(kind="bar",x="Fruits",stacked=True,ec="black")
plt.title("Stacked Bar Chart",style="italic",weight="bold")
plt.legend()
plt.xlabel("Fruits",style="oblique",weight="bold")
plt.ylabel("No.of Students",style="oblique",weight="bold")
plt.xticks(rotation=45,ha="right")
plt.show()
```
