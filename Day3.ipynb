# Z-Score Correlation Pairplots

```python
import numpy as np 
import pandas as pd 
```

```python
weights = [45,67,89,78,45,69,86,64,57,49,30,90,84]
df = pd.DataFrame(weights,columns = ['weight'])
df
```

```python
# z score = x- mu % sigma
```

```python
df.describe().T
```

### Using pandas

```python
s= pd.Series(weights)
print(s.head())
pmu = s.mean()
psig = s.std(ddof = 1)
print(pmu,psig)
s.std()
```

```python
mean = np.mean(df['weight'])
std_pop = np.std(df['weight'], ddof=0)   # population std
std_samp = np.std(df['weight'], ddof=1)  # sample std

print(std_samp,std_pop)
z_pop = (df['weight'] - mean) / std_pop
z_samp = (df['weight'] - mean) / std_samp
```

### Using Statistics

```python
import statistics as st
print(st.mean(weights),st.stdev(weights))
```

```python
mean = st.mean(df['weight'])
std_samp = st.stdev(df['weight'])   # sample std
std_pop = st.pstdev(df['weight'])   # population std
print(std_samp,std_pop)
```

### Using numpy

```python
mu = np.mean(df['weight'])
sigma = np.std(df['weight'])
print(mu,sigma)
```

```python
mean = np.mean(df['weight'])
std_pop = np.std(df['weight'], ddof=0)   # population std
std_samp = np.std(df['weight'], ddof=1)  # sample std

print(std_samp,std_pop )
z_pop = (df['weight'] - mean) / std_pop
z_samp = (df['weight'] - mean) / std_samp
```

```python
df['z_weight'] =( df['weight'] - mu )/ sigma
df
```
