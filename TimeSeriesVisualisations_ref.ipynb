#### Line plot

```python
# create a line plot
import pandas as pd
from matplotlib import pyplot as plt
```

```python
series = pd.read_csv('daily-minimum-temperatures.csv', \
                   parse_dates=[0] )   
series
```

```python
series.info()
```

```python
series = pd.read_csv('daily-minimum-temperatures.csv', \
                      index_col=0,parse_dates=[0])
series.head()
```

```python
series.info()
```

```python
# Plot kde using pandas
series.plot(kind='kde')
plt.show()
```

```python
import seaborn as sns
sns.displot(series, kde=True, stat = 'density')
```

```python
# Line plot
plt.figure(figsize=(20,6))
plt.plot(series)
```

#### Box and Whisker Plots by Interval

```python
# Box plot for all 10 years
plt.boxplot(series, vert = False)
```

```python
# Boxplot by year
fig,ax = plt.subplots(figsize=(20,5))
sns.boxplot(x = series.index.year, 
                y = series.Temp,
                ax = ax)
```

```python
series.index.month
```

```python
fig, ax = plt.subplots(figsize=(20,5))
sns.boxplot(x = series.index.month, 
                y = series.Temp, 
                ax = ax)
```

#### Lag plot

```python
# series = read_csv('https://raw.githubusercontent.com/slmsshk/DataSet/main/daily-minimum-temperatures.csv', header=0, index_col=0,parse_dates=True)

# create a lag scatter plot
from pandas.plotting import lag_plot

lag_plot(series, lag=10)
plt.show()
```

```python
# create an autocorrelation plot
from statsmodels.graphics.tsaplots import \
                               plot_acf, plot_pacf
plt.figure(figsize = (40,10))
plot_acf(series,lags=90)
plt.show()
```

```python
plot_pacf(series,lags=90)
```
