## **Different y-axis parameters on Histogram:**

---

### 1. **Count (default)**

* **What it shows**: The number of observations (data points) that fall into each bin.
* **Use case**: Most common form when you want to know "how many values" lie in a range.
* **Example**: If bin = \[10â€“20], and 7 values fall inside, the height of that bar = 7.

ðŸ“ˆ **Interpretation**: Bar heights directly represent sample size per bin.

---

### 2. **Probability / Relative Frequency**

* **What it shows**: The fraction of the total number of data points in each bin.
* **How itâ€™s calculated**:

  $$
  \text{Relative Frequency} = \frac{\text{Count in bin}}{\text{Total number of observations}}
  $$
* **Use case**: Useful when comparing histograms of datasets with different sizes.

ðŸ“ˆ **Interpretation**: The sum of all bar heights = 1.

---

### 3. **Density (normalized histogram)**

* **What it shows**: A scaled version of frequency such that the **total area under the histogram = 1**.
* **How itâ€™s calculated**:

  $$
  \text{Density in bin} = \frac{\text{Count in bin}}{N \cdot \text{Bin Width}}
  $$

  where $N$ = total number of data points.
* **Use case**: When you want to compare the histogram to a probability distribution curve (e.g., Normal distribution, KDE).

ðŸ“ˆ **Interpretation**: Useful in probability/statistics; bar heights depend on bin width.

---

### 4. **Probability Density Estimate (KDE curve on top)**

* Not strictly histogram y-axis, but often plotted alongside.
* It estimates the **probability density function (PDF)** of the data.
* Smooth curve, continuous counterpart of density histogram.

---


#### Code 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Generate sample data

np.random.seed(42)  # same data is reproduced when executed multiple times
data = np.random.normal(loc=50, scale=10, size=200)  # Generate synthetic data for visualization

# Create a DataFrame for easier handling
df = pd.DataFrame(data, columns=['value'])
df
```

```python
# Plot histogram with count y-axis parameter

plt.hist(df['value'], bins=10, color='skyblue', edgecolor='black')
plt.title("Histogram - Count (default)")
plt.ylabel("Count")
```

```python
# Plot histogram with probability as y-axis parameter
plt.hist(df['value'], bins=10, density=False, \
         weights=np.ones_like(df['value'])/len(df['value']), 
               color='lightgreen', edgecolor='black')
plt.title("Histogram - Probability (Relative Frequency)")
plt.ylabel("Probability")
```

```python
# Plot histogram with density as y-axis parameter
plt.hist(df['value'], bins=10, density=True, \
         color='salmon', edgecolor='black')
plt.title("Histogram - Density (Normalized)")
plt.ylabel("Density")
```

```python
# Plot histogram with density as y-axis and with KDE for comparison
import seaborn as sns
sns.histplot(df["value"], stat= 'density', kde = True)
plt.title("Histogram + KDE Overlay")
plt.ylabel("Density")
```
