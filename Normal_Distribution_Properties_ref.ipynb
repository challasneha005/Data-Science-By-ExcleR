### 1. Definition

The **Normal Distribution** (also called **Gaussian distribution**) is a **continuous probability distribution** that is symmetric around the mean.

* The **mean = median = mode**.
* The shape is a **bell curve**.
* Many natural and social phenomena (heights, exam scores, measurement errors) approximately follow this distribution.

---

### 2. Key Properties

* Symmetrical around the mean.
* Defined by **two parameters**:

  * Mean (Î¼): determines the **center** of the distribution.
  * Standard deviation (Ïƒ): determines the **spread (width)**.
* About **68%** of data falls within 1Ïƒ of the mean.
* About **95%** of data falls within 2Ïƒ of the mean.
* About **99.7%** of data falls within 3Ïƒ of the mean. (*Empirical Rule / 68-95-99.7 rule*).

---

### 3. Probability Density Function (PDF)

The formula for the normal distribution is:

$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{ -\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2 }
$$

Where:

* $\mu$ = Mean
* $\sigma$ = Standard Deviation
* $x$ = Variable

---

### 4. Shape and Relation to Skewness & Kurtosis

* A **perfect normal distribution** has **skewness = 0** and **kurtosis = 0** (excess kurtosis).
* If skewness â‰  0 â†’ distribution is asymmetric.
* If kurtosis â‰  0 â†’ distribution is either too flat or too peaked compared to normal.

---


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import seaborn as sns

# Generate normal distribution data
mu, sigma = 20, 5   # mean=20, std=5 (standard normal)
data = np.random.normal(mu, sigma, 1000)

# Plot histogram + PDF curve

sns.histplot(data, kde = True)
plt.title("Normal Distribution (Î¼=20, Ïƒ=)5")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```


### ðŸ”¹ Skewness

* **Definition:**
  Skewness measures the **asymmetry** of the probability distribution of a dataset.
* **Types:**

  * **Positive Skew (Right-Skewed):** Long tail on the right. Mean > Median.
  * **Negative Skew (Left-Skewed):** Long tail on the left. Mean < Median.
  * **Zero Skew:** Distribution is symmetric (like Normal).

---

### ðŸ”¹ Kurtosis

* **Definition:**
  Kurtosis measures the **tailedness** (peakedness) of a distribution.
* **Types:**

  * **Mesokurtic (â‰ˆ 3):** Normal distribution.
  * **Leptokurtic (> 3):** Heavy tails, sharp peak, possible outliers
  * **Platykurtic (< 3):** Light tails, flat peak, less possibility for outliers

ðŸ‘‰ In Python (`pandas`/`scipy.stats`), kurtosis is usually returned as **excess kurtosis**:

* Normal distribution â†’ 0
* Positive â†’ Leptokurtic
* Negative â†’ Platykurtic
---

```python
# Computation of Skewness and Kurtosis for Iris Dataset

import pandas as pd
from sklearn.datasets import load_iris

# Load iris dataset from sklearn library
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Compute skewness and kurtosis
print("=== Skewness ===")
print(df.skew())   # Pandas method

print("\n=== Kurtosis ===")
print(df.kurt())   # Pandas method (excess kurtosis)
```
