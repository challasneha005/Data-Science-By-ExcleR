```python
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold, \
                                         cross_val_predict
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report
```

```python
# Load dataset
df = pd.read_csv("dia_clean.csv")
df
```

```python
# Drop index column if unnecessary
X = df.drop(columns=['Unnamed: 0', 'class'])
y = df['class']
```

```python
# Define model
gb_clf = GradientBoostingClassifier(random_state=42)

# Stratified K-Fold
skf = StratifiedKFold(n_splits=5, shuffle=True, \
                           random_state=42)
# Cross-validated predictions
y_pred = cross_val_predict(gb_clf, X, y, cv=skf)
```

```python
# Classification report
report = classification_report(y, y_pred, \
                               output_dict=False)
print(report)
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np

# cross_val_score: get fold-wise scores (accuracy here)
scores = cross_val_score(gb_clf, X, y, cv=skf, scoring='accuracy')

# Mean and std of accuracy
mean_acc = np.mean(scores)
std_acc = np.std(scores)

scores, mean_acc, std_acc
```
