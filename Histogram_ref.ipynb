### 1. Definition

A **Histogram** is a graphical representation of the distribution of a dataset. It groups **continuous numerical data** into **intervals (bins)** and displays the frequency (count) of values that fall into each bin as a bar.

---

### 2. Key Features

* Unlike a bar chart, a histogram is for **continuous data**.
* The **x-axis** represents intervals (bins).
* The **y-axis** represents frequency (number of data points in each bin).
* Bars are **adjacent with no gaps**, since data is continuous.

---

### 3. Difference between Histogram and Bar Chart

| Feature           | Histogram                    | Bar Chart            |
| ----------------- | ---------------------------- | -------------------- |
| Data type         | Continuous numerical         | Categorical          |
| Gaps between bars | No (bars touch)              | Yes (bars separated) |
| X-axis            | Intervals (ranges of values) | Categories (labels)  |

---

### 4. When to Use

* To understand the **distribution** of data.
* To detect **skewness** (left/right).
* To spot **outliers**.
* To check if data is approximately **normal (bell-shaped)**.

---

### 5. Example Dataset

Suppose exam scores of 20 students are:

```
45, 50, 52, 53, 55, 56, 57, 58, 60, 62, 
65, 66, 68, 70, 72, 74, 75, 78, 80, 85
```

If we divide into bins of width 10:

* 40â€“49 â†’ 1 student
* 50â€“59 â†’ 7 students
* 60â€“69 â†’ 5 students
* 70â€“79 â†’ 5 students
* 80â€“89 â†’ 2 students

The histogram shows the **frequency distribution** across these ranges.

---

### 6. Advantages

* Shows overall shape of the data distribution.
* Easy to identify skewness, spread, and outliers.
* Helps in selecting statistical models (e.g., normal distribution).

---


### Using Matplotlib

````python
import matplotlib.pyplot as plt

data = [45, 50, 52, 53, 55, 56, 57, 58, 60, 62, 
        65, 66, 68, 70, 72, 74, 75, 78, 80, 85]

plt.hist(data, bins=5, color="skyblue", edgecolor="black")
plt.title("Histogram of Exam Scores")
plt.xlabel("Score Range")
plt.ylabel("Frequency")
plt.show()
````

### Using Pandas

````python
import pandas as pd
import matplotlib.pyplot as plt

# Example dataset
data = [45, 50, 52, 53, 55, 56, 57, 58, 60, 62, 
        65, 66, 68, 70, 72, 74, 75, 78, 80, 85]

# Convert to Pandas Series
s = pd.Series(data)

# Plot histogram
s.hist(bins=5, color="lightgreen", edgecolor="black", grid =False)

plt.title("Histogram of Exam Scores (Pandas)")
plt.xlabel("Score Range")
plt.ylabel("Frequency")
plt.show()
````

### Using Seaborn

````python
import seaborn as sns

data = [45, 50, 52, 53, 55, 56, 57, 58, 60, 62, 
        65, 66, 68, 70, 72, 74, 75, 78, 80, 85]

sns.histplot(data, bins=5, color="skyblue", edgecolor="black", kde=True)
plt.title("Histogram of Exam Scores")
plt.xlabel("Score Range")
plt.ylabel("Frequency")
plt.show()
````
