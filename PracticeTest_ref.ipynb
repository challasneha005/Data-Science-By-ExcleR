#### **Observations**
1. The data type for all the columns sepal.length, sepal.width, petal.length, petal.width is continuous numerical value and therefore we prefer to plot histogram
2. For the Variety column the data type is object and there are only 3 categories. Hence Bar chart is preferred.
3. There are no missing values, hence we can go for plotting the visualizations

```python
# Obtain descriptive statistics (Transposed for better understanding)
df.describe().T
```

#### Histogram and kde plot for Sepal Length

```python
# Histogram and kde plot for sepal.length
import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot(x= "sepal.length", data = df, kde=True)
plt.title("Histogram for Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

#### Inference
1. The kde curve appears to be slightly right-skewed
2. It is also confirmed from the data points spread from its mean of 5.843  with min value of 4.3 and max value of 7.9. So data is more spread towards maximum than towars minimum

#### Histogram and kde plot for Sepal width

```python
# Histogram and kde plot for sepal.length
import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot(x= "sepal.width", data = df, kde=True)
plt.title("Histogram for Sepal Width")
plt.xlabel("Sepal Widthth")
plt.ylabel("Frequency")
plt.show()
```

#### Inference
1. The kde curve appears to be almost symmetrical with slight stretch towards right
2. It is also confirmed from the data points spread from its mean of 3.05, with min value of 2 and max value of 4.4. So data is almost symmetrical from its mean 

```python
# Histogram and kde plot for petal.length

import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot(x= "petal.length", data = df, kde=True)
plt.title("Histogram for Petal Length")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.show()
```

#### Inference
1. The kde curve appears to be  asymmetrical with bimodal nature (two peaks) with skewed to left

```python
# Histogram and kde plot for petal.length

import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot(x= "petal.width", data = df, kde=True)
plt.title("Histogram for Petal Width")
plt.xlabel("Petal Width")
plt.ylabel("Frequency")
plt.show()
```

#### Bar Chart for Variety column

```python
# Bar Chart 
counts = df["variety"].value_counts()
sns.barplot(x = counts.index, y = counts.values, hue = counts.index)
plt.title("Distribution of Iris Flower Variety")
plt.ylabel = "Counts"
```

#### Inference:
1. There are three categories each with equal number of data points. Indicates a balanced classes
