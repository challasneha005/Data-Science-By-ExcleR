#### Numpy Arrays
- NumPy arrays offer powerful library for numerical computing in Python. 
- Homogeneous Data Type: NumPy arrays contain elements that are all of the same type.
- Multidimensional: Arrays can have multiple dimensions
- Slicing and Indexing: Advanced slicing and indexing features 
- Efficient Memory Usage: NumPy arrays have a fixed size at creation.
- NumPy arrays are a foundational tool for scientific computing in Python.

#### 1D arrays

```python
a = np.array([10,20,30,40])
print(a)
print(type(a))
```

```python
# Reshapig an array using reshape()
a = np.array([10,20,30,40])
b = a.reshape(2,2)
print(b)
print(b.shape)
```

```python
# Use of around()
d = np.array([1.3467, 3.10987,4.91236])
print(d)
np.around(d,2)
```

```python
# Use of np.sqrt()
d = np.array([1.3467, 3.10987,4.912])
print(d)
print(np.around(np.sqrt(d),2))
```

```python
# Mathematical operations on rows and cols
a2 =np.array([[3,4,6],[7,9,10],[4,6,12]])
a2
```

```python
print(a2.sum(axis = 1))
print(a2.sum(axis=0))
```

```python
# Matrix operations
a3 = np.array([[3,4,5],[7,2,8],[9,1,6]])
a3
```

```python
# Matrix operations
a3 = np.array([[3,4,5],[7,2,8],[9,1,6]])
print(a3)
np.fill_diagonal(a3,0)
print(a3)
```

```python
# Define two matrices and multiply them
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# Perform matrix multiplication
C = np.matmul(A, B)

print(C)
```

```python
# Print the Transpose of the matrices
print(A.T)
print(B.T)
```
