# Splitting data

```python
Train = Walmart.head(147)
Test = Walmart.tail(12)
```

```python
Test
```

```python
#Linear Model
import statsmodels.formula.api as smf 

linear_model = smf.ols('Footfalls~t',data=Train).fit()
pred_linear =  pd.Series(linear_model.predict(pd.DataFrame(Test['t'])))
rmse_linear = np.sqrt(np.mean((np.array(Test['Footfalls'])-np.array(pred_linear))**2))
rmse_linear
```

```python
#Exponential

Exp = smf.ols('log_footfalls~t',data=Train).fit()
pred_Exp = pd.Series(Exp.predict(pd.DataFrame(Test['t'])))
rmse_Exp = np.sqrt(np.mean((np.array(Test['Footfalls'])-np.array(np.exp(pred_Exp)))**2))
rmse_Exp
```

```python
#Quadratic 

Quad = smf.ols('Footfalls~t+t_square',data=Train).fit()
pred_Quad = pd.Series(Quad.predict(Test[["t","t_square"]]))
rmse_Quad = np.sqrt(np.mean((np.array(Test['Footfalls'])-np.array(pred_Quad))**2))
rmse_Quad
```

```python
#Additive seasonality 

add_sea = smf.ols('Footfalls~Jan+Feb+Mar+Apr+May+Jun+Jul+Aug+Sep+Oct+Nov',data=Train).fit()
pred_add_sea = pd.Series(add_sea.predict(Test[['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov']]))
rmse_add_sea = np.sqrt(np.mean((np.array(Test['Footfalls'])-np.array(pred_add_sea))**2))
rmse_add_sea
```

```python
#Additive Seasonality Quadratic 

add_sea_Quad = smf.ols('Footfalls~t+t_square+Jan+Feb+Mar+Apr+May+Jun+Jul+Aug+Sep+Oct+Nov',data=Train).fit()
pred_add_sea_quad = pd.Series(add_sea_Quad.predict(Test[['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','t','t_square']]))
rmse_add_sea_quad = np.sqrt(np.mean((np.array(Test['Footfalls'])-np.array(pred_add_sea_quad))**2))
rmse_add_sea_quad
```

```python
predictions = pd.DataFrame(pred_add_sea_quad, columns=["predicted"])
predictions
```

```python
predictions.index
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,4))
plt.plot(Train["Footfalls"])
plt.plot(Test["Footfalls"])
plt.plot(predictions["predicted"])
plt.show()
```

### Forecasting in to future by creating future dataframe

```python

# Assume Train data goes till t = 159 (Dec 2012 for example)
last_t = 159
future_periods = 12  # Forecast next 12 months

# Create future dataframe
future = pd.DataFrame({
    't': range(last_t + 1, last_t + future_periods + 1)
})
future['t_square'] = future['t'] ** 2

# Create month column sequence
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

# Assign month names cyclically
future['month'] = [(months[(i-1) % 12]) for i in future['t']]

# Create dummies
future = pd.get_dummies(future, columns=['month'])
print(future)
```

```python
# Make sure it has the same dummy columns as Train (e.g., drop Dec if you dropped it earlier to avoid dummy trap)
for col in ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov']:
    if col not in future.columns:
        future[col] = 0

# Predict
future['forecast'] = add_sea_Quad.predict(future)

print(future[['t','forecast']])
```

```python
# 3. Plot
plt.figure(figsize=(12,6))
plt.plot(Train['t'], Train['Footfalls'], label="Train", color='blue')
plt.plot(Test['t'], Test['Footfalls'], label="Test", color='green')
#plt.plot(Train['t'], Train['predicted'], '--', label="Train Predicted", color='skyblue')
plt.plot(predictions.index, predictions['predicted'], '--', label="Test Predicted", color='lime')
plt.plot(future['t'], future['forecast'], label="Future Forecast", color='red')

plt.xlabel("Time (t)")
plt.ylabel("Footfalls")
plt.title("Walmart Footfalls Forecast using Multiple Linear Regression")
plt.legend()
plt.grid(True)
plt.show()
```
