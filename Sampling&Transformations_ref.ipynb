#### Upsampling Data

```python
# upsample to daily intervals
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
series = pd.read_csv('sales.csv', header=0, \
                index_col='Month', parse_dates=[0])
```

```python
series
```

```python
# Plot the line plot
plt.figure(figsize=(12,4))
plt.plot(series)
```

```python
# Do upsampling using resample 'D'
sub_series = series.head(2)
upsampled = sub_series.resample('D').mean()
print(upsampled)
```

```python
# Perform linear interpolation

interpolated = upsampled.interpolate(method='linear')
print(interpolated)
interpolated.info()
```

```python
plt.figure(figsize=(12,4))
plt.plot(interpolated)
```

#### Downsampling Data

```python
# Perform downsampling
resample = series.resample('Q')
quarterly_mean_sales = resample.mean()
```

```python
plt.figure(figsize=(12,4))
plt.plot(series)
plt.plot(quarterly_mean_sales)
# quarterly_mean_sales.plot()
```

# Tranformations

```python
#series = read_csv('https://raw.githubusercontent.com/slmsshk/DataSet/main/airline-passengers.csv', header=0, index_col=0, parse_dates=True)
series = pd.read_csv("airline-passengers.csv", index_col=[0],\
                                   parse_dates=[0]  )
series
```

```python
# Line plot
plt.figure(figsize=(12,4))
plt.plot(series["Passengers"])
```

#### Square Root Transform

```python
L1 = series["Passengers"].to_list()
```

```python
series['Pass_sqrt'] = np.sqrt(L1)
```

```python
series
```

```python
# Line Plot passengers, Pass_sqrt
plt.figure(figsize=(12,4))
plt.plot(series["Passengers"])
plt.plot(series["Pass_sqrt"])
```

```python
# line plots as sub plots
plt.subplot(211)
plt.plot(series['Passengers'])

plt.subplot(212)
plt.plot(series['Pass_sqrt'])
```

#### Log Transform

```python
# Apply log Transform

series["Pass_log"] = np.log(L1)
series
```

```python
# line plots as sub plots for log transform
plt.subplot(211)
plt.plot(series['Passengers'])

plt.subplot(212)
plt.plot(series['Pass_log'])
```
