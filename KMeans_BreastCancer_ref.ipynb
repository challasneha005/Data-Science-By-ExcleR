```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
from sklearn.cluster import KMeans
```

**Clustering- Divide the Breast Cancer data in to groups (Clusters)**

```python
df = pd.read_csv("breast_cancer.csv")
df
```

```python
df.isna().sum()
```

#### Standardization of the data

```python
# Read all numeric columns in to df1
df1 = df.iloc[:,1:]
df1.head()
```

```python
cols = df1.columns

# Standardisation function 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_df1 = pd.DataFrame(scaler.fit_transform(df1),\
                              columns = cols )
scaled_df1
```

#### Finding optimal k value using elbow plot

```python
# Build 2 clusters using KMeans Cluster algorithm
from sklearn.cluster import KMeans
clusters_new = KMeans(2, random_state=0) # Specify 3 clusters
clusters_new.fit(scaled_df1)
```

```python
# print the cluster labels
clusters_new.labels_
```

```python
set(clusters_new.labels_)
```

```python
#Assign clusters to the Univ data set
df1['clusterid_new'] = clusters_new.labels_
```

```python
df1
```

#### Find the quality of clusters

```python
# Quality of clusters is expressed in terms 
# of Silhoutte score

from sklearn.metrics import silhouette_score
score =silhouette_score(scaled_df1,clusters_new.labels_ , \
                        metric='euclidean')
score
```
