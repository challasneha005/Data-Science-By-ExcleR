### 1. Definition

A **Bar Chart** is a graphical representation of categorical data using rectangular bars. The **length (or height)** of each bar is proportional to the value (frequency, count, or percentage) it represents.

---

### 2. Key Features

* **Categories** are displayed on one axis (x-axis for vertical bars, y-axis for horizontal bars).
* **Values** are displayed on the other axis.
* Bars are **uniform in width** but vary in height/length.
* Spaces between bars highlight that the data is **categorical** (not continuous).

---

### 3. Types of Bar Charts

1. **Vertical Bar Chart** â€“ Bars stand upright. Most common.
2. **Horizontal Bar Chart** â€“ Useful when category names are long.
3. **Grouped Bar Chart** â€“ Compares subcategories within each main category.
4. **Stacked Bar Chart** â€“ Subcategories stacked within each bar, showing part-to-whole relationships.

---

### 4. When to Use

* To compare **frequencies** of different categories.
* To highlight **differences** between groups.
* To show **trends** across categories (but not time series).

---

### Bar Chart Code Using Matplotlib

```python
import pandas as pd

# Example dataset
fruits = ["Apple", "Banana", "Orange", "Grapes"]
students = [10, 8, 7, 5]

df = pd.DataFrame({"Fruit": fruits, "Students": students})
df
```

```python
import matplotlib.pyplot as plt

plt.bar(fruits, students, color="skyblue", edgecolor="black")
plt.title("Favorite Fruits - Vertical Bar Chart")
plt.xlabel("Fruit")
plt.ylabel("Number of Students")
plt.show()
```

### Using Pandas

```python
df2 = pd.DataFrame({
    "Fruit": ["Apple", "Banana", "Orange", "Grapes"],
    "Girls": [6, 4, 3, 2],
    "Boys": [4, 4, 4, 3]
})
print(df2)

df2.plot(kind="bar", x="Fruit", edgecolor="black")
plt.title("Favorite Fruits - Grouped Bar Chart")
plt.ylabel("Number of Students")

# Rotate x-axis labels
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python

df2.plot(kind="bar", x="Fruit", stacked=True, edgecolor="black")
plt.title("Favorite Fruits - Stacked Bar Chart")
plt.ylabel("Number of Students")
plt.show()
```
