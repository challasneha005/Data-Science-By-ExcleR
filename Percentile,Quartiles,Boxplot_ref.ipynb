
# ðŸ“˜ Percentiles in Statistics

### ðŸ”¹ Definition

A **percentile** is a measure that indicates the value below which a given percentage of data in a dataset falls.

* For example, the **25th percentile (P25)** is the value below which **25% of the data** lie.
* The **50th percentile (P50)** is the **median** of the dataset.
* The **90th percentile (P90)** is the value below which **90% of the data** lie.

Percentiles are widely used in **education, health, finance, and data analysis** to understand the relative standing of a value within a dataset.

---

### ðŸ”¹ How to Compute Percentiles (Manual Steps)

1. **Arrange the data** in ascending order.
2. **Find the rank (position) using the formula:**

$$
R = \frac{P}{100} \times (N+1)
$$

Where:

* $P$ = desired percentile (e.g., 25, 50, 90)
* $N$ = number of data points
* $R$ = rank (position in ordered data)

3. **Interpret the rank (R):**

   * If $R$ is an integer â†’ the percentile is the value at that position.
   * If $R$ is not an integer â†’ interpolate between the two closest values.

---

### ðŸ”¹ Example (Manual Calculation)

Dataset (scores of 10 students):

$$
[45, 50, 52, 53, 55, 56, 57, 58, 60, 62]
$$

* **Step 1:** Already sorted.
* **Step 2:** Find the 25th percentile (P25).

$$
R = \frac{25}{100} \times (10+1) = 2.75
$$

* **Step 3:** The 25th percentile lies between the **2nd and 3rd values**.

  * 2nd value = 50, 3rd value = 52
  * Interpolation:

$$
P25 = 50 + 0.75 \times (52 - 50) = 51.5
$$

âœ… The 25th percentile = **51.5**

---

* Percentiles are useful to **compare relative performance** (e.g., exam scores, health indicators).

---



# ðŸ“˜ Quartiles in Statistics

### ðŸ”¹ Definition

**Quartiles** are values that divide a dataset into **four equal parts**, each containing **25% of the observations** when the data is arranged in ascending order.

They are essentially **special cases of percentiles**:

* **Q1 (First Quartile)** = 25th percentile (P25)
  â†’ 25% of the data lies below Q1.
* **Q2 (Second Quartile / Median)** = 50th percentile (P50)
  â†’ 50% of the data lies below Q2.
* **Q3 (Third Quartile)** = 75th percentile (P75)
  â†’ 75% of the data lies below Q3.

---

### ðŸ”¹ Interquartile Range (IQR)

The **IQR** is a measure of statistical dispersion and is defined as:

$$
IQR = Q3 - Q1
$$

* It shows the spread of the **middle 50% of the data**.
* Useful in detecting **outliers** (values outside $[Q1 - 1.5 \times IQR, Q3 + 1.5 \times IQR]$).

---

### ðŸ”¹ Example (Manual Calculation)

Dataset (scores):

$$
[45, 50, 52, 53, 55, 56, 57, 58, 60, 62]
$$

* **Q1 (25th percentile):** Already computed = **51.5**
* **Q2 (Median, 50th percentile):** Middle value between 5th (55) and 6th (56) = **55.5**
* **Q3 (75th percentile):**

$$
R = \frac{75}{100} \times (10+1) = 8.25
$$

So Q3 lies between 8th value (58) and 9th value (60).

$$
Q3 = 58 + 0.25 \times (60 - 58) = 58.5
$$

âœ… Quartiles: Q1 = 51.5, Q2 = 55.5, Q3 = 58.5
âœ… IQR = Q3 - Q1 = 58.5 - 51.5 = **7.0**

---


#### Python Code

```python
import numpy as np
import pandas as pd

# Example dataset
data = [45, 50, 52, 53, 55, 56, 57, 58, 60, 62]

# -------------------------
# Using NumPy
# -------------------------
print("=== Using NumPy ===")
# Percentiles
p25 = np.percentile(data, 25)
p50 = np.percentile(data, 50)   # median
p75 = np.percentile(data, 75)
p90 = np.percentile(data, 90)

print("25th Percentile (P25):", p25)
print("50th Percentile (P50 / Median):", p50)
print("75th Percentile (P75):", p75)
print("90th Percentile (P90):", p90)
```

```python
# -------------------------
# Using Pandas
# -------------------------
print("\n=== Using Pandas ===")
series = pd.Series(data)

# Percentiles
print("25th Percentile (P25):", series.quantile(0.25))
print("50th Percentile (P50 / Median):", series.quantile(0.50))
print("75th Percentile (P75):", series.quantile(0.75))
print("90th Percentile (P90):", series.quantile(0.90))

# Quartiles
Q1 = series.quantile(0.25)
Q2 = series.quantile(0.50)
Q3 = series.quantile(0.75)
IQR = Q3 - Q1
print("Quartiles -> Q1:", Q1, "Q2:", Q2, "Q3:", Q3)
print("Interquartile Range (IQR):", IQR)
```
