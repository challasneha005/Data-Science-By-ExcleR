```python
import os
from dotenv import load_dotenv
from huggingface_hub import InferenceClient

#load token from 
load_dotenv()
hf_token = os.getenv("HF token")

model_id = "stabilityai/stable-diffusion-xl-base-1.0"

client = InferenceClient(
    model=model_id,
    token=hf_token
)

prompt = "A futuristic city at sunset, ultra realistic, 4k"

image = client.text_to_image(prompt)

image.save("output.png")
print("Image saved as output.png")
```

```python
from dataclasses import dataclass
from typing import List
```

```python
@dataclass
class Expense:
    data:str
    amount:float
    category:str
class ExpenseAgent:
    def __init__(self):
        self.expenses:List[Expense] =[]
    def add_expense(self,date,amount,category):
        self.expenses.append(Expense(date,amount,category))
    def summarize_expenses(self):
        summary = {}
        for exp in self.expenses:
            if exp.category not in summary:
                summary[exp.category]=0
            summary[exp.category] +=exp.amount
        print("Monthly expense summary:")
        for cat,total in summary.items():
            print(f"{cat}: ${total:.2f}")
        return summary
```

```python
agent = ExpenseAgent()
agent.add_expense("2024-01-05",50.00,"Food")
agent.add_expense("2024-01-10",25.00,"Transport")
agent.add_expense("2024-01-15",100.00,"Food")

summary = agent.summarize_expenses()
summary
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np
docs =  ["LLM accuracy can be improved using better data quality and fine-tuning.",
    "Prompt engineering helps improve LLM responses.",
    "Vector databases are used for efficient retrieval in RAG systems.",
    "Evaluation metrics are important for measuring LLM performance."
        ]

model = SentenceTransformer("all-MiniLM-L6-v2")
embs = model.encode(docs)

def retrieve(query):
    q_emb = model.encode([query])[0]
    sims = np.dot(embs,q_emb)
    idx = np.argmax(sims)
    return docs[idx]

print(retrieve("How do I improve LLM Accuracy?"))
```
