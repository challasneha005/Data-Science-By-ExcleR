# STATISTICS

## MEASURES OF CENTRAL TENDENCY

## Using Numpy

```python
import numpy as np
heights = np.array([167,182,156,178,171,159,164,164])
print(np.mean(heights))
np.median(heights)
#no mode
```

```python
#mean shifted if outliers
heights = np.array([167,182,156,178,171,159,164,164,1780])
print(np.mean(heights))
np.median(heights)
```

## Using Statistics

```python
import statistics as st
heights = np.array([167,182,156,178,171,159,164,164])
print("Mean",st.mean(heights))
print("Median",st.median(heights))
print("Mode:",st.mode(heights))
```

```python
#bi and multi model mode
heights = np.array([167,182,156,178,171,159,164,164,167])
print("Mean",st.mean(heights))
print("Median",st.median(heights))
print("Mode:",st.mode(heights))
```

## Using Pandas

```python
import pandas as pd
heights = np.array([167,182,156,178,171,159,164,164,167])
df = pd.Series(heights)
print("Mean",df.mean())
print("Median",df.median())
print("Mode:",df.mode())
```

## METHODS OF DISPERSION

## Using Numpy

```python
import numpy as np
heights = np.array([167,182,156,178,171,159,164,164,167])
print("population std:", np.std(heights))
print("sample std:", np.std(heights,ddof = 1)) # delta degrees of freedom sample 1,population 0
```

```python
#range
np.ptp(heights)
```

```python
np.max(heights) - np.min(heights)
```

```python
print(np.var(heights))
np.var(heights,ddof =1)
```

## Using Statistics


```python
import statistics as st
heights = np.array([167,182,156,178,171,159,164,164])
print("Sample variance:",st.variance(heights))
print("Population variance:",st.pvariance(heights))
print("Standard deviation:",st.stdev(heights))
```

## Using Pandas

```python
import pandas as pd
heights = np.array([167,182,156,178,171,159,164,164,167])
df = pd.Series(heights)
print("Sample Variance:",df.var()) #sample
print("Sample std:",df.std()) #sample
print("Population Variance:",df.var(ddof=0)) # population
print("Population std:",df.std(ddof=0)) # population
```

# CHARTS

```python
fruits = ['Apple','Banana','Orange','Grapes']
students = [10,8,7,5]
df = pd.DataFrame({'Fruits':fruits,'Students':students})
df
```

## BAR CHART

```python
import matplotlib.pyplot as plt
colors = ['skyblue','orange','pink','green']
#plt.bar(fruits,students,color = 'skyblue',edgecolor = 'black')
plt.bar(fruits,students,color= colors,edgecolor = 'black')
plt.title("Bar Chart - Vertical")
plt.xlabel("Fruits ")
plt.ylabel("Students num")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.barh(fruits,students,color = ['skyblue','orange','pink','green'],edgecolor = 'black')
plt.title("Bar Chart")
plt.xlabel("Students num")
plt.ylabel("Fruits - Vertical")
plt.show()
```

### Using Pandas

```python
df2 = pd.DataFrame({
    "Fruits":['Apple','Banana','Orange','Grapes'],
    'Girls':[6,4,3,2],
    'Boys':[4,4,4,3]
})
df2
```

### Grouped Barchart

```python
df2.plot(kind = 'bar',x='Fruits',edgecolor = 'black')
plt.title("Fruits Fav")
plt.ylabel('Student Num')
plt.show()
```

###  Stacked Barchart

```python
df2.plot(kind = 'bar',x='Fruits',stacked=True,edgecolor = 'black')
plt.title("Fruits Fav")
plt.ylabel('Student Num')
plt.show()
```

### Using Seaborn 

```python
import seaborn as sns

sns.barplot(x="Fruits", y="Girls", data=df2, edgecolor="black")
plt.show()
# sns.barplot(x = fruits.index, y = fruits.values, edgecolor = "black", hue=fruits.index)
```

```python
df2_melted = df2.melt(id_vars="Fruits", value_vars=["Girls", "Boys"],
                      var_name="Gender", value_name="Count")

sns.barplot(x="Fruits", y="Count", hue="Gender", data=df2_melted, edgecolor="black")
plt.show()
```

## HISTOGRAM 

### Using Matplotlib

```python
import matplotlib.pyplot as plt
data = [45,50,52,53,55,56,57,58,60,62,65,66,68,70,72,74,75,78,80,85]
plt.hist(data,bins = 5,color = "skyblue",edgecolor='black')
plt.title("histogram")
plt.xlabel("Score Range")
plt.ylabel("Frequency")
plt.show()
```

### Using Pandas

```python
s = pd.Series(data)
s.hist(bins = 5,color = "skyblue",edgecolor='black')
plt.title("histogram")
plt.xlabel("Score Range")
plt.ylabel("Frequency")
plt.show()
```

### Using Seaborn 

```python
import seaborn as sns
sns.histplot(data,bins = 9,color = "skyblue",edgecolor='black',kde=True) #kernel density estimate
plt.title("histogram")
plt.xlabel("Score Range")
plt.ylabel("Frequency")
plt.show()
```
