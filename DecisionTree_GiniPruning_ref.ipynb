### Pruning of Decision Trees

### Pruning in Decision Trees

Pruning is a technique used in decision tree algorithms to reduce the size of the tree and prevent overfitting. Overfitting occurs when a model learns the noise in the training data rather than the actual underlying patterns, leading to poor generalization on new, unseen data. Pruning simplifies the decision tree by removing sections of the tree that provide little to no power in predicting target variables.

There are two main types of pruning:
1. **Pre-pruning (Early Stopping)**: Stop the growth of the tree early by setting constraints such as maximum depth, minimum samples per leaf, or minimum information gain.
2. **Post-pruning (Pruning after Tree Construction)**: Allow the tree to grow to its maximum size and then remove nodes or subtrees that do not contribute significantly to the model’s performance.

### Cost Complexity Pruning (CCP) Alpha Method

Cost Complexity Pruning, also known as the **CCP Alpha** method, is a post-pruning technique used to trim a decision tree to find the optimal size of the tree that balances model complexity and prediction accuracy. It works by penalizing the complexity of the tree (i.e., the number of terminal nodes) to avoid overfitting.

**In the decision tree, when each subtree is examined, if the reduction in classification error is small compared to the penalty from having many terminal nodes, then that subtree is likely to be pruned.**

#### Steps Involved in CCP Alpha Pruning:

1. **Compute Cost Complexity for Each Subtree**:
   - For each subtree $$ T_t $$ of the original tree $$ T $$, compute the cost complexity as:
     $$
     R_\alpha(T_t) = R(T_t) + \alpha \cdot |T_t|
     $$
     where:
     - $$ R(T_t) $$ is the misclassification rate or error of the subtree $$ T_t $$.
     - alpha is the complexity parameter that penalizes the number of terminal nodes (leaves) $$ |T_t|$$.

2. **Identify the Optimal alpha**:
   - Find the value of alpha that results in the subtree with the lowest cost complexity. This is done by cross-validation, where multiple values of alpha are tested, and the one that minimizes the cross-validated error is chosen.

3. **Prune the Tree**:
   - Prune the tree by removing nodes that do not contribute significantly to the model’s performance based on the optimal alpha.

1. **Cost Complexity Pruning Path**: `clf.cost_complexity_pruning_path(X_train, y_train)` computes the effective alphas and the corresponding total leaf impurities (cost complexities) for pruning.
2. **Train Multiple Trees**: For each alpha value, a decision tree is trained and stored.
3. **Select Best Tree**: Based on cross-validation or another validation technique, select the best tree that balances accuracy and complexity.
4. **Plot and Evaluate**: Visualize the pruned tree and evaluate its performance on the test set.

By using the CCP Alpha method, you can create a decision tree that generalizes better to new data by avoiding overfitting.

#### Finding best value of ccp_alpha for pruning

```python
# List the array of alphas
path = model.cost_complexity_pruning_path(x_train,y_train)
alphas = path["ccp_alphas"]
alphas
```

```python
# Selection of optimal ccp_alpha
train_accuracy, test_accuracy = [],[]
for each in alphas:
    model = DecisionTreeClassifier(criterion = 'gini',ccp_alpha=each)
    model.fit(x_train,y_train)
    train_predict = model.predict(x_train)
    test_predict = model.predict(x_test)
    train_accuracy.append(accuracy_score(y_train,train_predict))
    test_accuracy.append(accuracy_score(y_test,test_predict))
    
```

```python
plt.plot(alphas,train_accuracy, label = "train accuracy")
plt.plot(alphas,test_accuracy, label = "test accuracy")
plt.xlabel("ccp_alpha")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()
```

```python
# Prediction on train data

pred_test = model.predict(x_test)
print(classification_report(y_test,pred_test))
```
