### Recommender Systems
A recommender system in machine learning is a type of system that aims to predict the preferences or interests of users and suggests items that might be of interest to them. These items can range from movies, music, and books to more specific recommendations like products on an e-commerce website or articles on a news site. The primary goal of a recommender system is to help users discover products or services that they may not have come across otherwise.

Recommender systems are broadly classified into three main types:

1. **Content-Based Filtering:**
   - This type of system recommends items based on the features of the items and a profile of the user's preferences. The system tries to recommend products that are similar to those that a user has liked in the past. For example, if a user has shown an interest in mystery novels, the system might recommend other books in the mystery genre.

2. **Collaborative Filtering:**
   - Collaborative filtering recommends items by identifying patterns of similarity between users and/or items based on their interaction history. There are two main approaches:
     - **User-based collaborative filtering:** This method finds users who have similar preferences and tastes as a given user and recommends items that these similar users have liked.
     - **Item-based collaborative filtering:** This method recommends items that are similar to items the user has liked in the past, based on the similarity of items as judged by ratings from other users.

3. **Hybrid Recommender Systems:**
   - Hybrid systems combine both content-based and collaborative filtering approaches in an effort to improve the accuracy of recommendations and to overcome the limitations inherent in each approach. For example, a hybrid system might use collaborative filtering to find similar users but also use content-based filtering to ensure that the items recommended are close to the user’s content preferences.

Each type of recommender system has its strengths and weaknesses, and the choice of which type to use often depends on the specific needs of the application and the nature of the data available.

### Metrics used to find similarity

In collaborative filtering recommender systems, the choice of similarity metrics is crucial for determining how closely items or users match one another. Here are some commonly used similarity metrics:

1. **Cosine Similarity:**
   - This measures the cosine of the angle between two vectors in a multi-dimensional space. In the context of collaborative filtering, these vectors could be users or items with their dimensions being ratings. Cosine similarity is particularly useful when the magnitude of the vector (i.e., the quantity of ratings) is not as important as the angle between them.
   
   $${Cosine Similarity} (\mathbf{A}, \mathbf{B}) = \frac{\mathbf{A} \cdot \mathbf{B}}{\|\mathbf{A}\| \|\mathbf{B}\|}$$


2. **Pearson Correlation Coefficient:**
   - Pearson correlation measures the linear correlation between two variables, normalized by their standard deviations and means. In user-based collaborative filtering, it's used to compute the correlation between different users’ rating behaviors, focusing on ratings co-variance from the users' average ratings. This can help correct for users who are consistently high or low raters.
$${Pearson Correlation} (\mathbf{X}, \mathbf{Y}) = \frac{\sum (X_i - \overline{X})(Y_i - \overline{Y})}{\sqrt{\sum (X_i - \overline{X})^2 \sum (Y_i - \overline{Y})^2}}$$

3. **Euclidean Distance:**
   - This is the geometric distance in the multi-dimensional space. For recommender systems, it can be used to find users with similar preferences by measuring the shortest path between points (user rating vectors). It’s more sensitive to variations in vector magnitude, which can be both a strength and a weakness depending on the context.
   $${Euclidean Distance} (\mathbf{A}, \mathbf{B}) = \sqrt{\sum_{i=1}^n (A_i - B_i)^2}$$
4. **Jaccard Index:**
   - This metric is used for cases where the data is binary (such as "like" or "dislike", "watched" or "not watched"). It measures the size of the intersection divided by the size of the union of the sample sets. It is useful in scenarios where the absence of a rating is as meaningful as its presence.
   $${Jaccard Index} (A, B) = \frac{|A \cap B|}{|A \cup B|}$$

5. **Spearman's Rank Correlation:**
   - Similar to Pearson but instead uses rank values. It assesses how well the relationship between two variables can be described using a monotonic function. It’s useful in cases where the data doesn't necessarily meet the assumptions required for Pearson correlation but still requires a measure of directional association.
$${Spearman's Rank Correlation} (\mathbf{X}, \mathbf{Y}) = \frac{\sum (R(X_i) - \overline{R(X)})(R(Y_i) - \overline{R(Y)})}{\sqrt{\sum (R(X_i) - \overline{R(X)})^2 \sum (R(Y_i) - \overline{R(Y)})^2}}$$

6. **Adjusted Cosine Similarity:**
   - This is an adaptation of cosine similarity where item ratings are adjusted by subtracting the user’s average rating before calculating the cosine of the angle between these adjusted vectors. It's particularly useful in item-based collaborative filtering, helping to normalize user ratings behaviors across different scales.
$${Adjusted Cosine Similarity} (\mathbf{X}, \mathbf{Y}) = \frac{\sum (X_i - \overline{X_u})(Y_i - \overline{Y_u})}{\sqrt{\sum (X_i - \overline{X_u})^2 \sum (Y_i - \overline{Y_u})^2}}$$

Each of these metrics serves a different purpose and may be more or less appropriate depending on the specific characteristics of the data set and the goals of the recommender system. In practice, these metrics can also be tested and tuned to optimize performance.

### Complete code for generating recommendations

```python
# Function that generates recommendations
def recommendations(x,similar_users):
    recommended_movies = []
    for i in similar_users:
        user_1=movies_df[movies_df['userId']== x]
        user_2=movies_df[movies_df['userId']== i]
        movies_watched_by_user1 = list(set(user_1.movie))
        movies_watched_by_user2 = list(set(user_2.movie))
        
        for movie_name in movies_watched_by_user2:
            if movie_name not in movies_watched_by_user1:
                recommended_movies.append(movie_name)
    return set(recommended_movies)
            
```

```python
# Scan for the first five row indexes to generate recommendations

i = 0
for x in user_sim_df.index:
    similar_users = user_sim_df[user_sim_df[x]>=0.8].index
    print("Recommendations for user: ",x)
    movies = recommendations(x,similar_users)
    print(movies)
    i = i+1
    if i > 4:
        break
    
```
