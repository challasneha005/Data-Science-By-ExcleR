# CONFIDENCE INTERVAL

```python
import numpy as np
interval  = 0.95  # 1 - alpha
alpha = 1 - 0.95
mu = 1990
sigma = 2500
n = 140
zfactor = stats.norm.ppf(1- alpha/2)
print("zfactor:",zfactor)
confidenceIntervalLeft = mu - zfactor * (sigma / np.sqrt(n))
confidenceIntervalRight = mu + zfactor * (sigma / np.sqrt(n))
print(confidenceIntervalLeft,confidenceIntervalRight)
```

### z distribution

```python
n = 100
mu = 40
sigma = 10
c = 0.95
alpha = 0.05
zfactor = stats.norm.ppf(1- alpha/2)
print("zfactor:",zfactor)
confidenceIntervalLeft = mu - zfactor * (sigma / np.sqrt(n))
confidenceIntervalRight = mu + zfactor * (sigma / np.sqrt(n))
print(confidenceIntervalLeft,confidenceIntervalRight)
```

```python
n = 100
mu = 40
sigma = 10
c = 0.99
alpha = 0.01
zfactor = stats.norm.ppf(1- alpha/2)
print("zfactor:",zfactor)
confidenceIntervalLeft = mu - zfactor * (sigma / np.sqrt(n))
confidenceIntervalRight = mu + zfactor * (sigma / np.sqrt(n))
print(confidenceIntervalLeft,confidenceIntervalRight)
```

### t distribution

```python
n = 100
mu = 40
sigma = 10
c = 0.95
alpha = 0.05
tfactor = stats.t.ppf(1- alpha/2,df = n-1)
print("tfactor:",tfactor)
confidenceIntervalLeft = mu - tfactor * (sigma / np.sqrt(n))
confidenceIntervalRight = mu + tfactor * (sigma / np.sqrt(n))
print(confidenceIntervalLeft,confidenceIntervalRight)
```

```python
## 99
n = 100
mu = 40
sigma = 10
c = 0.99
alpha = 0.01
tfactor = stats.t.ppf(1- alpha/2,df = n-1)
print("tfactor:",tfactor)
confidenceIntervalLeft = mu - tfactor * (sigma / np.sqrt(n))
confidenceIntervalRight = mu + tfactor * (sigma / np.sqrt(n))
print(confidenceIntervalLeft,confidenceIntervalRight)
```

# HYPOTHESIS TESTING 

### IceCream Problem
### h0 : <0.3
### h1 : >=0.3

```python
# p<=0.05 to reject ho ,so h0 accept
# p>0.05 fail to reject ho ,so h1 accept
```

```python
import numpy as np
import pandas as pd
from scipy import stats
sample = np.array([0.593,0.142,0.329,0.691,0.231,0.793,0.519,0.392,0.418])
n = 9
x_bar = np.mean(sample)
std = np.std(sample,ddof = 1)
mu = 0.3
t_stat = (x_bar-mu) / (std/np.sqrt(n))
p = 1-stats.t.cdf(t_stat,n-1)
x_bar,std,t_stat,p
```

```python
if p<=0.05:
    print("reject H0")
else:
    print("Accept H0")
```

### Bakery Problem left tailed
### mu >= 30
### mu < 30

```python
import numpy as np
import pandas as pd
from scipy import stats
n = 40
mu = 30
x_bar = 28
sigma = 5
z_stat = (x_bar- mu) / (sigma/np.sqrt(n))
p = stats.norm.cdf(z_stat)
p
```

```python
if p<=0.05:
    print("reject H0")
else:
    print("Accept H0")
```

### 2 Tailed test

```python
mu = 4
sigma = 3
n =50
x_bar = 5.8
z = (x_bar-mu) / (sigma/np.sqrt(n))
z
```

```python
p = 2*(1-stats.norm.cdf(z))
p
```

```python
if p<=0.05:
    print("reject H0")
else:
    print("Accept H0")
```
