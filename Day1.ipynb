## Use for loop print even list and oddlist (do not use comprehension)

```python
# List l1 is defined here
L1 = [100,80,99,204,77,109,56,89]
even = []
odd = []
for i in range(0,8):
    if(L1[i]%2 ==0):
        even.append(L1[i]) 
    else:
        odd.append(L1[i])
print("Even list:",even)
print("Odd list:",odd)
```

```python
# comprehension
L1 = [100,80,99,204,77,109,56,89]
even = []
odd = []
for i in L1:(even.append(i) if(i%2 == 0)  else odd.append(i))
print("Even list:",even)
print("Odd list:",odd)
```

## Write program to search for a string in a tuple and print its presence in it

```python
tup1 = ("Hello", "Hello World", "Hello friend", "Hello Sir");
#search for "Hello Sir"
# membership operator
if ("Hello Sir" in tup1):
    print("present")
else:
    print("Not Present")
```

```python
# for loop
for i in tup1:
    if i == "Hello Sir" :
        flag = 1
    else:
        flag = 0
if flag:
    print("present")
else:
    print("Not Present")
```

```python
for i in tup1:
    if i == "Helloo Sir" :
        print("present")
        break
else:
    print("Not Present")
 
```

## Take the list L4  Use for loop to print numbers less <100 with continue

```python
L4 = [100,80,99,204,77,109,56,89]
l = []
for i in L4:
    if i >= 100:
        continue
    else:
        l.append(i)
l
```
